```python
# Plot 
import pylab
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


sns.set(style="white")
sns.set_context("paper", font_scale = 2.5)

total = 2
fig, axn = plt.subplots(1, total, figsize = (20,20),sharey=True)
nums= [0,1]

for i, ax in enumerate(axn.flat):
    col = colonies[str(nums[i])]
    tensions = [e.ground_truth for e in col.tot_edges]
    mean_ten = np.mean(tensions)
    tensions = [e/mean_ten for e in tensions]

    col.plot_tensions(ax, fig, tensions, min_x=450, max_x=550, min_y=430, max_y=550, 
                      min_ten = 0, max_ten = 3, specify_color = 'jet',cbar = 'no', lw = 3)
    plt.setp(ax.get_yticklabels(), visible=False)
    plt.setp(ax.get_xticklabels(), visible=False)
    ax.set(xlim = [400,650], ylim = [400,650], aspect = 1)

pylab.savefig('ground_truth.pdf')
```

```python
from DLITE.ManualTracingMultiple import ManualTracingMultiple
colonies_numbers = [int(k) for k, v in colonies.items()]
ManualTracingMultipleInstance = ManualTracingMultiple(colonies_numbers)

colonies_numbers2 = [int(k) for k, v in colonies2.items()]
ManualTracingMultipleInstance2 = ManualTracingMultiple(colonies_numbers2)
```

```python
# We first find labels of edges that are present in all the colonies
common_edge_labels = ManualTracingMultipleInstance.get_repeat_edge(colonies)
common_cell_labels = ManualTracingMultipleInstance.get_repeat_cell(colonies)

common_edge_labels2 = ManualTracingMultipleInstance.get_repeat_edge(colonies2)
common_cell_labels2 = ManualTracingMultipleInstance.get_repeat_cell(colonies2)
```

```python
# Make the dataframes
edges_dataframe, cells_dataframe = ManualTracingMultipleInstance.seaborn_plot(None, colonies,
                                                                                  common_edge_labels,
                                                                                  common_cell_labels,
                                                                                  ground_truth = True)
# nodes_dataframe = ManualTracingMultipleInstance.seaborn_nodes_dataframe(colonies, None)

# Make the dataframes
edges_dataframe2, cells_dataframe2 = ManualTracingMultipleInstance.seaborn_plot(None, colonies2,
                                                                                  common_edge_labels2,
                                                                                  common_cell_labels2,
                                                                                  ground_truth = True)
#nodes_dataframe2 = ManualTracingMultipleInstance.seaborn_nodes_dataframe(colonies2, None)
```

```python
tension_table = edges_dataframe.pivot_table(
        values='Local_normalized_tensions', 
        index=['Edge_Labels'], 
        columns='Time')

ground_truth_tension_table = edges_dataframe.pivot_table(
        values='Ground_truth', 
        index=['Edge_Labels'], 
        columns='Time')

tension_table2 = edges_dataframe2.pivot_table(
        values='Local_normalized_tensions', 
        index=['Edge_Labels'], 
        columns='Time')
```

```python
import seaborn as sns
import pylab
sns.set_context('poster')
fig, axn = plt.subplots(1, 3, figsize = (8,4),sharey=True)
cbar_ax = fig.add_axes([1.1, .15, .01, .7], title = 'Tension')

for i, ax in enumerate(axn.flat):
    if i == 0:
        df = ground_truth_tension_table
        title = 'Ground_truth'
    elif i == 1:
        df = tension_table
        title = 'CellFIT'
    elif i == 2:
        df = tension_table2
        title = 'DLITE'
    ax.set_title(title)
    sns.heatmap(df, ax=ax,
                cbar=i == 0, vmin=0, vmax=1, cmap = 'jet',
                cbar_ax=None if i else cbar_ax)


pylab.savefig('heatmap.pdf')
```

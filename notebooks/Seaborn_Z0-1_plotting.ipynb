```python


# ZO-1
pkl_file = open('Post_surface_evolver_small_colony_v1_KKT.pkl', 'rb')
#pkl_file = open('post_surface_evolver_small_colony_v3_KKT.pkl', 'rb')
#pkl_file = open('post_surface_evolver_small_colony_v4_KKT.pkl', 'rb')
#pkl_file = open('Post_surface_evolver_small_colony_v2_KKT.pkl', 'rb')

# old working
#pkl_file = open('80_150_tensions_0.1_2_KKT.pkl', 'rb')
#pkl_file = open('surface_evolver_dec_edge_groups_80_150_KKT.pkl', 'rb')

colonies_2 = pickle.load(pkl_file)
#colonies_2 = dict((k, v) for k, v in colonies_2.items() if 0 <= int(k) < 18)

common_edge_labels = test.get_repeat_edge(colonies_2)
common_cell_labels = test.get_repeat_cell(colonies_2)
edges_dataframe_2, cells_dataframe_2 = test.seaborn_plot(None, colonies_2, common_edge_labels,common_cell_labels, ground_truth= False)
nodes_dataframe_2 = test.seaborn_nodes_dataframe(colonies_2, None)
tensor_dataframe_2 = test.seaborn_cells_dataframe_tensor(colonies_2)
```

```python


aa = edges_dataframe_1.pivot_table(
        values='Local_normalized_tensions', 
        index=['Edge_Labels'], 
        columns='Time')

b = edges_dataframe_2.pivot_table(
        values='Local_normalized_tensions', 
        index=['Edge_Labels'], 
        columns='Time')
```

```python

test_uncon = edges_dataframe_1.pivot_table(
        values='Stochasticity_in_tension', 
        index=['Edge_Labels'], 
        columns='Time')

test_cellfit = edges_dataframe_2.pivot_table(
        values='Stochasticity_in_tension', 
        index=['Edge_Labels'], 
        columns='Time')
```

```python
sns.set_context("paper", font_scale=3)    
fig, axn = plt.subplots(1, 2, figsize = (8,10),sharey=True)
cbar_ax = fig.add_axes([1.00, .15, .01, .69], title = 'Tension')

for i, ax in enumerate(axn.flat):
    if i == 0:
        df = b
        title = 'CELLFIT'
    else:
        df = aa
        title = 'Local optimization'
    ax.set_title(title)
    sns.heatmap(df, ax=ax,
                cbar=i == 0,
                vmin=0, vmax=3,cmap = 'jet',
                cbar_ax=None if i else cbar_ax)

# cbar_ax.set_label('Tension')

pylab.savefig('4B.pdf', bbox_inches='tight')
pylab.savefig('4B.svg', bbox_inches='tight')
```

```python
sns.set_context("paper", font_scale=3)    
fig, axn = plt.subplots(1, 2, figsize = (8,10),sharey=True)
cbar_ax = fig.add_axes([1.02, .15, .01, .69], title = '$\Delta$Tension')

for i, ax in enumerate(axn.flat):
    if i == 0:
        df = test_cellfit
        title = 'CELLFIT'
    else:
        df = test_uncon
        title = 'Local optimization'
    ax.set_title(title)
    sns.heatmap(df, ax=ax,
                cbar=i == 0,
                vmin=-2, vmax=2,cmap = 'jet',
                cbar_ax=None if i else cbar_ax)

# cbar_ax.set_label('Tension')

pylab.savefig('4B.pdf', bbox_inches='tight')
pylab.savefig('4B.svg', bbox_inches='tight')
```

```python
check_data = {'Label': [], 'Time': [], 'Method': [],'Stochasticity_in_tension': [], 'Tension': []}

for idx, series in edges_dataframe_1.iterrows():
    check_data['Label'].append(idx[0])
    check_data['Time'].append(idx[1])
    check_data['Tension'].append(series['Local_normalized_tensions'])
    check_data['Method'].append('Local optimization')
    check_data['Stochasticity_in_tension'].append(series['Stochasticity_in_tension'])

# for idx, series in edges_dataframe_1.iterrows():
#     check_data['Label'].append(idx[0])
#     check_data['Time'].append(idx[1])
#     check_data['Tension'].append(series['Ground_truth'])
#     check_data['Ground_truth'].append(series['Ground_truth'])
    
#     check_data['Ground_truth_error'].append((series['Ground_truth'] - series['Ground_truth']) )
#     check_data['Method'].append('Ground_truth')
#     check_data['Stochasticity_in_tension'].append(series['Ground_truth_stochasticity_in_tension'])
    
for idx, series in edges_dataframe_2.iterrows():
    check_data['Label'].append(idx[0])
    check_data['Time'].append(idx[1])
    check_data['Tension'].append(series['Local_normalized_tensions'])
        
    check_data['Method'].append('CELLFIT')
    check_data['Stochasticity_in_tension'].append(series['Stochasticity_in_tension'])
```

```python
import pandas as pd

two_method_tension_for_hue = pd.DataFrame(check_data)
```

```python
sns.set(style="darkgrid")
sns.set_context("paper", font_scale=3.6)

fig, ax = plt.subplots(1,1,figsize = (26,6))
g = sns.violinplot(ax = ax, x = 'Time', y = 'Stochasticity_in_tension', hue = 'Method', data = two_method_tension_for_hue)
g.set( ylim = [-5,6])
ax.set_ylabel('$\Delta$Tension')
ax.set_xlabel('Time')

plt.legend(loc='upper left')
pylab.savefig('7E.pdf', bbox_inches = 'tight')
pylab.savefig('7E.svg', bbox_inches = 'tight')


```

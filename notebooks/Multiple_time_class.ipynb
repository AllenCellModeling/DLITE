```python
# import packages

import pickle, pprint
import numpy as np
import matplotlib.pyplot as plt
import collections
import sys
sys.path.insert(0, '../cell_soap/')
from cell_describe import node, edge, cell, colony, data, data_multiple
```

# Load data into data_multiple class

```python
# data_multiple(data)
pkl_file = open('zo1_graph_timelapse_sample_v2.pkl','rb')
data1 = pickle.load(pkl_file)
pkl = data_multiple(data1)
```

# Compute results

```python
#  Choose time and cutoff
t = 23
cutoff = 1
```

```python
# 5 outputs - colony, tensions, pressures, tension-matrix, pressure-matrix
# pkl.compute(time, cutoff)
pkl.compute(t, cutoff)
```

# Plot results

```python
# pkl.plot only plots tensions for now because pressures seem singular

fig, ax = plt.subplots(1,1, figsize = (8,5))
# pkl.plot(ax, fig, time, cutoff)
pkl.plot(ax, fig, t, cutoff, lw = 20)
```

```python
# Plot cells 
fig, ax1 = plt.subplots(1,1, figsize = (8,5))
#pkl.plot_cells(ax, t, cutoff)
pkl.plot_cells(ax1, t, cutoff)
```

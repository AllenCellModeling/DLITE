```python
# Split the txt file at ','
# Good example, T4 with cleanup 13
# T7 - missing edge
with open('cellfit_synthetic_trace.txt','r') as f:
    a = [l.split(',') for l in f]
```

```python
# Get x and y co-ordinates
import numpy as np
x,y, X, Y = [],[], [],[]

for num in a:
    if len(num) == 2:
        x.append(int(num[0]))
        y.append(int(num[1].strip('\n')))
    if len(num) == 1:
        X.append(x)
        Y.append(y)
        x = []
        y = []
X.append(x)
Y.append(y)
X.pop(0)
Y.pop(0)
```

```python
# plot stuff
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1)
ax.set(xlim =[-100,1100], ylim = [-100,1300], aspect = 1)

for i in range(len(X)):
    ax.plot(X[i], Y[i], lw = 4)
    ax.plot(X[i][0], Y[i][0], 'ok')
    ax.plot(X[i][-1], Y[i][-1], 'ok')

plt.show()
```

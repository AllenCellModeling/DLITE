```python
import numpy as np
import numpy.linalg as la
def py_ang(v1, v2):
    """ Returns the angle in radians between vectors 'v1' and 'v2'    """
    cosang = np.dot(v1, v2)
    sinang = la.norm(np.cross(v1, v2))
    return np.rad2deg(np.arctan2(sinang, cosang))
```

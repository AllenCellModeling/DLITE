# Create synthetic graph to test cell finding on

```python
import numpy as np
import matplotlib.pyplot as plt 
import networkx
import random
```

With our imports done, we'll create the graph. This is a triangular graph where we then select a set of random edges to remove, thus converting some cells from 3-node-cells to 4- or 5-node-cells.

Since there is an element of randomness to the next cell, you may need to run it a couple of times to get a network that looks like a good fit to the cell description. 

```python
graph = networkx.generators.lattice.triangular_lattice_graph(5,8)
to_remove = random.sample(list(graph.edges), 7)
graph.remove_edges_from(to_remove)
pos = networkx.spring_layout(graph)
networkx.draw(graph, pos)
```

When we calculated the spring layout, we got xy positions for each of those nodes.
